# Matplotlib

Matplotlib is a very usefull library to make graphs. There are muliple types of graphs available where you can configure axes, ticks, gridline, legends and more details.

You can save the graph as pdf or svg file, so when you zoom in the graph will still be neat. This is not the case with other types of image files that are defined by pixels.

You can make graphs in a jupyter notebook and display it there which is very useful.

Also, Pandas uses matplotlib to make the graphs.

There are a lot of multiple options available to create your graphs. As always, ocumentation will be your friend to create the graphs you need for your projects.

````python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
````

Let's take a look at the tutorial:

https://matplotlib.org/stable/users/explain/quick_start.html#figure

Most methods will also parse a string-indexable object like a dict, a structured numpy array, or a pandas.DataFrame.

Matplotlib allows you to provide the data keyword argument and generate plots passing the corresponding strings for x and y variables.

````python
np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50), # Return evenly spaced values within a given interval
        'c': np.random.randint(0, 50, 50), # random integers from 0 (inclusive) to 50 (exclusive)
        'd': np.random.randn(50)} # Return a sample from the standard normal distribution
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', c='c', s='d', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
````

## Styling Options

https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

## Colors
https://matplotlib.org/stable/users/explain/colors/colors.html#colors-def

````python
# Colors
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.scatter(data1, data2, s=50, facecolor='C0', edgecolor='k')
````

## Axes labels and text

````python
# Axes labels and text
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

# data histogram
n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Aardvark lengths\n (not really)')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
````

## Legends

````python
# legends
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.plot(np.arange(len(data1)), data1, label='data1')
ax.plot(np.arange(len(data2)), data2, label='data2')
ax.plot(np.arange(len(data3)), data3, 'd', label='data3')
ax.legend()
````

## Customizing ticks

````python
# customizing ticks
xdata = np.arange(len(data1))

fig, axs = plt.subplots(2, 1, layout='constrained')
axs[0].plot(xdata, data1)
axs[0].set_title('Automatic ticks')

axs[1].plot(xdata, data1)
axs[1].set_xticks(np.arange(0, 100, 30), ['zero', '30', 'sixty', '90'])
axs[1].set_yticks([-1.5, 0, 1.5])  # note that we don't need to specify labels
axs[1].set_title('Manual ticks')
````

## Additional axis

````python
# Additional axis
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)

fig, (ax1, ax3) = plt.subplots(1, 2, figsize=(7, 2.7), layout='constrained')
l1, = ax1.plot(t, s)
ax2 = ax1.twinx()
l2, = ax2.plot(t, range(len(t)), 'C1')
ax2.legend([l1, l2], ['Sine (left)', 'Straight (right)'])

ax3.plot(t, s)
ax3.set_xlabel('Angle [rad]')
ax4 = ax3.secondary_xaxis('top', functions=(np.rad2deg, np.deg2rad))
ax4.set_xlabel('Angle []')
````
